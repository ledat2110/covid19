```python
import covsirphy as cvp
import pandas as pd
import numpy as np
```

```python
country = 'Italy'
```

```python
data = pd.read_csv('./covid_19_jhu_data_web_scrap_and_cleaning/full_grouped.csv')
data = data[data['Country/Region'] == country].reset_index()
data = data[data['Confirmed'] > 0]
```

```python
data = data.fillna(0)
```

```python
df = pd.read_csv('./population_2020.csv')
```

```python
population = df[df['Country (or dependency)'] == country]['Population (2020)']
```

```python
population = int(population)
print(population)
```

```python
num_sample = len(data)
```

```python
population = [population for _ in range(num_sample)]
```

```python
data['Population'] = population
```

```python
data.tail()
```

```python
data['Susceptiple'] = data['Population'] - data['Confirmed']
```

```python
data['Log susceptiple'] = np.log(data['Susceptiple'])
```

```python
data.tail()
```

```python
susceptiple = np.asarray(data['Susceptiple'])
```

```python
new_susceptiple = susceptiple[:-1] - susceptiple[1:]
```

```python
new_susceptiple = np.insert(new_susceptiple, 0, 0)
```

```python
data['Decresed susceptiple'] = new_susceptiple
```

```python
data = data.drop(columns=['WHO Region', 'Country/Region', 'index'])
```
